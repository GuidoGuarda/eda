```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Criando um boxplot
sns.boxplot(data=df_musica)

# Exibindo o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

media_estilos = df_musica.mean()


# Verificar se 'media_estilos' contém valores válidos
print(media_estilos)

# Criando o gráfico de pizza
plt.figure(figsize=(8, 6))
plt.pie(media_estilos, labels=media_estilos.index, autopct='%1.1f', startangle=140)

# Adicionando um título
plt.title('Percentual de Médias de Votos por Estilo Musical')

# Exibindo o gráfico
plt.show()

```
